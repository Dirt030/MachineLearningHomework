```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

## Import File

```python
data = pd.read_csv('winequality-red.csv');data
```

## Data Details

```python
data.info()
```

### 1- Distribution

```python
plt.figure(figsize=(25, 12))
for i, idx in enumerate(data.columns):
    fig = plt.subplot(3, 4, 1 + i)
    data[idx].plot.kde()
    plt.title(idx)
```

### 1-Quality Distribution

```python
plt.figure(figsize=(25, 12))
for i, idx in enumerate(data.columns):
    fig = plt.subplot(3, 4, 1 +i)
    sns.barplot(x = 'quality', y = idx, data = data)
```

### Correlation Matrix

```python
plt.figure(figsize=(25, 20))

sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
```
